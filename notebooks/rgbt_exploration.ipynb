# RGB-T Data Exploration\n
Inspect synchronized RGB and thermal samples and verify pairing quality.

```python
from pathlib import Path\n
import matplotlib.pyplot as plt\n
id
from PIL import Image\n
\n
root = Path('../data/sample_rgbt')\n
split = 'train'\n
rgb_files = sorted((root / 'rgb' / split).glob('*.jpg')) + sorted((root / 'rgb' / split).glob('*.png'))\n
thermal_files = {p.stem: p for p in (root / 'thermal' / split).glob('*') if p.suffix.lower() in {'.jpg', '.jpeg', '.png'}}\n
print('RGB samples:', len(rgb_files))
```

```python
if rgb_files:\n
    p = rgb_files[0]\n
    t = thermal_files[p.stem]\n
    rgb = Image.open(p).convert('RGB')\n
    th = Image.open(t).convert('L')\n
\n
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))\n
    ax[0].imshow(rgb)\n
    ax[0].set_title('RGB')\n
    ax[0].axis('off')\n
\n
    ax[1].imshow(th, cmap='inferno')\n
    ax[1].set_title('Thermal')\n
    ax[1].axis('off')\n
    plt.tight_layout()
```
